$$ [T]_{\beta,\alpha}=
\begin{bmatrix}
1 & 2 & -1\\
1 & 0 & -1\\
0 & 1 & 0
\end{bmatrix} $$
e $$ [G]_{\alpha,\beta}=
\begin{bmatrix}
1 & 1 & 2\\

\end{bmatrix}. $$





(a) Determine bases para $Ker\left(  G\circ T\right)  $

(b) Considere agora $\gamma = \{1, 1+x, 1+x^2\}$ base de $\mathcal{P}%
_{2}\left(  \mathbb{R}\right)$. Quais são as coordenadas de $p(x)= 6+2x+3x^2$ na base $\gamma$ ?


#### Solução

```python
T = np.array([[1, 2, -1],
            [1, 0, -1], 
            [0, 1, 0]])

G = np.array([[1, 1, 2],
            [1, -1, 0],
            [-1, 1, 2]])

G_compos_T = G @ T
print(G_compos_T)

# Fazendo X = [x, y, z] e verificando quando G_compos_T @ X = 0 temos:
# Base do Ker(G o T) = {1 + x²}

# Encontrando a imagem
im_1 = G_compos_T @ np.array([1, 0, 0])
im_2 = G_compos_T @ np.array([0, 1, 0])

print(im_1) # p(x) = 1
print(im_2) # p(x) = 4 + 2x
# Base da Im(G o T) = {1, 4 + 2x}

# Para encontrar o valor do polinomio na base gama = {1, 1 + x, 1 + x²}, primeiro fazemos a transformação com G_compos_T
# Depois encontramos uma matriz mudança de base da canônica para gama, e efetuamos a multiplicação
```
