# Exercício 1

Use o Teorema Espectral e encontre uma trasformação linear cujos autovetores são  $v_1=[1,1]$  e  $v_2=[1,−1]$  e os autovalores correspondentes são  $\alpha_1$ e  $\alpha_2$ , respectivamente. Verifique graficamente o efeito desta transformação no círculo  $x_2+y_2=1$, quando: 

a) $\alpha_1 = 0.5$ e $\alpha_2 = 0.1$.

c) $\alpha_1 = 0.5$ e $\alpha_2 = 0.5$.

Comente os resultdos obtidos.

**Solução exercícico 1**

O Teorema Espectral, diz que:

Toda matriz simétrica $T$ pode ser decomposta em função de outras duas matrizes, uma matriz invertível $P$ e uma matriz diagonal $D$. Mais específicamente, essa decomposição é da forma:

$$
A = P D P^{-1}
$$

Mas o que isso significa?

Sempre que uma matriz $A$ tiver autovetores suficientes para gerar todo o espaço, é possível representar essa transformação linear como um alongamento de uma base formada pelos autovetores. Essa nova transformação linear chamada de "alongamento" é representada pela matriz diagonal $D$, enquanto a mudança de base (da base original) para a base composta de autovetores, é formada pela multiplicação de matrizes $P$ e $P^{-1}$.

De forma visual, podemos representar essa decomposição de forma:

$$
A =
\underbrace{
\begin{bmatrix}
| & & | \\
v_1 & \cdots & v_k \\
| & & | \\
\end{bmatrix}
}_{Autovetores}
\cdot
\underbrace{
\begin{bmatrix}
\lambda_1 & 0 & 0 \\
0 & \ddots &  0\\
0 & 0 & \lambda_l \\
\end{bmatrix}
}_{Autovalores}
\cdot
\begin{bmatrix}
| & & | \\
v_1 & \cdots & v_k \\
| & & | \\
\end{bmatrix} ^{-1}
$$

E a transformação linear de um vetor qualquer $\mathbb{X} = [x_1, ..., x_k]$ é interpretado da forma:

$$
A
\cdot
\begin{bmatrix}
x_1 \\
\vdots \\
x_k \\
\end{bmatrix}
=
\underbrace{
\begin{bmatrix}
| & & | \\
v_1 & \cdots & v_k \\
| & & | \\
\end{bmatrix}
\cdot
\underbrace{
\begin{bmatrix}
\lambda_1 & 0 & 0 \\
0 & \ddots &  0\\
0 & 0 & \lambda_l \\
\end{bmatrix}
\cdot
\underbrace{
\begin{bmatrix}
| & & | \\
v_1 & \cdots & v_k \\
| & & | \\
\end{bmatrix} ^{-1}
\cdot
\begin{bmatrix}
x_1 \\
\vdots \\
x_k \\
\end{bmatrix}
}_{Base\,canônica\rightarrow Base\,autovetores}
}_{Alongamento}
}_{Base\,autovetores\rightarrow Base\,canônica}
$$


E uma matriz simétrica sempre possui autovetores suficientes para satisfazer representar todo o espaço, por isso o Teorema Espectral requer essas condições. Porém é necessário deixar claro que não são somente as matrizes simétricas que possuem autovetores suficientes para gerar todo o espaço da transformação linear, por isso podemos encontrar transformações lineares não simétricas mas que respeitam essa decomposição.

Da mesma forma que é possível representar a matriz $A$ utilizando seus autovetores e autovalores através do Teorema Espectral, também é possível realizar o caminho inverso e construir a transformação $A$ somente a partir de autovetores ($v_i$'s) e autovalores($\lambda_i$'s), e isso garante que a matriz formada $A$ terá autovetores $v_i$'s e autovalores $\lambda_i$'s.

```python
# Montando a transformação linear (uma para cada ítem)

"""
Primeiro é armazenado os autovetores v_1 e v_2 que serão compartilhados pelas transformações lineares dos itens a), b) e c).
Posteriormente, para cada item é armazenados seus respectivos autovalores e a matriz A_i é calculada utilizando o Teorema Espectral apontado
"""

# Auto vetores
autovetores = np.array([[1, 1],
                        [1, -1]])

# a)
autovalores = np.array([0.5, 0.1])                                      # Armazenando os autovalores (alfa_1 e alfa_2)
A1 = autovetores @ np.diag(autovalores) @ np.linalg.inv(autovetores)    # Utilizando o Teorema Espectral para formar a matriz A1 com os autovalores e autovetores enunciados

# b)
autovalores = np.array([0.1, 0.5])
A2 = autovetores @ np.diag(autovalores) @ np.linalg.inv(autovetores)

# c)
autovalores = np.array([0.5, 0.5])
A3 = autovetores @ np.diag(autovalores) @ np.linalg.inv(autovetores)

# Gerando pontos no ciruclo
n = 100                         # numero de pontos sobre o circulo
t = np.linspace(0,2*np.pi,n)
X = np.zeros((2,n))
X[0,:] = np.cos(t)
X[1,:] = np.sin(t)

# Encontrando a transformacao de X
A1X = A1 @ X
A2X = A2 @ X
A3X = A3 @ X

# Plotando os gráficos de cada transformação linear.
fig = plt.figure()

ax1 = fig.add_subplot(141)
ax1.set_aspect('equal')
ax1.plot(X[0,:],X[1,:])

ax2 = fig.add_subplot(142)
ax2.set_aspect('equal')
ax2.plot(A1X[0,:],A1X[1,:])

ax3 = fig.add_subplot(143)
ax3.set_aspect('equal')
ax3.plot(A2X[0,:],A2X[1,:])

ax4 = fig.add_subplot(144)
ax4.set_aspect('equal')
ax4.plot(A3X[0,:],A3X[1,:])

plt.show()
```

b) Faça um teste para verificar que a base obtida é de fato ortonormal.



**Solução exercício 2**

O processo de orgonalização de Gram-Schmidt é um processo utilizado para decompor um conjunto de vetores que formam a base de um espaço euclidiano, em uma base ortogonal. Além disso, os vetores podem ser normalizados posteriormente a fim formar uma base ortonormal.

A imagem abaixo demonstra as etapas do processo de Gram-Schmidt em uma base do $\mathbb{R}^{3}$ de forma animada:

![Animação Gram-Schmidt](https://upload.wikimedia.org/wikipedia/commons/e/ee/Gram-Schmidt_orthonormalization_process.gif)

Para formalização matemática, o processo de Gram-Schmidt é feito por meio de projeções dos vetores da base ($v_1, v_2, ..., v_k$) em vetores ortogonais ($u_1, u_2, ..., u_k$) e funciona da seguinte forma:

$ u_1 = v_1$

$ u_2 = v_2 - proj_{u_1}(v_2)$

$ u_3 = v_3 - proj_{u_1}(v_3) - proj_{u_2}(v_3)$

$ u_4 = v_4 - proj_{u_1}(v_4) - proj_{u_2}(v_4) - proj_{u_3}(v_4)$

$ \qquad \vdots$

$ u_k = v_k - \sum_{j=1}^{k-1} proj_{u_j}(v_k) $

Dessa forma, a cada etapa o próximo vetor formado vai ser necessáriamente ortogonal aos dois vetores $u_i$'s formados anteriormente. Então $\mathbb{U} = \{u_1, u_2, u_3, ..., u_k \}$ forma uma base ortogonal.

Para obter uma base ortonormal, é necessário dividir cada vetor de $\mathbb{U}$ pelo seu tamanho (sua norma), de forma que $\mathbb{E} = \{\frac{u_1}{\lVert u_1 \rVert}, \frac{u_2}{\lVert u_2 \rVert}, \frac{u_3}{\lVert u_3 \rVert}, ..., \frac{u_k}{\lVert u_k \rVert}\} = \{e_1, e_2, e_3, ..., e_k\}$

```python
def CGS(X):
    """
    CGS: Classical Gram-Schmidt
    Ortogonalizar os vetores armazenados nas colunas de X.
    Usando o algoritmo de Gram-Schmidt mostrado acima.
    
    Depois de computar os vetores orgononais, eles são normalizados.
    """
    V = X.copy()
    Q = np.zeros(X.shape)
    for j in range(X.shape[1]):
        vj = V[:,j]
        for k in range(j):
            vj = vj - np.dot(X[:,j], Q[:,k]) * Q[:,k]
        qj = vj / np.linalg.norm(vj)
        Q[:,j] = qj
    return Q


V = np.random.randint(0,20,size=(20,10))

"""
Checando dependencia linear entre os vetores coluna de V.
Referencia: Theo Bendit (https://math.stackexchange.com/users/248286/theo-bendit), Linear independence of vectors that form non-square matrix, URL (version: 2017-08-31): https://math.stackexchange.com/q/2411838
"""
while np.linalg.det(V.T @ V) == 0:
    print(V)
    print('Os vetores colunas de V são linearmente dependentes. Dessa forma não constitui base')
    print('Gerando novos vetores aleatórios para a base')
    V = np.random.randint(0,20,size=(20,10))

print(V)


"""
Computando os vetores ortonormais em U utilizando o algoritmo classico de Gram-Schmidt
Depois, é computado a diferença entre o vetor gerado e os produtos internos entre as colunas
de U. O resultado ideal seria uma matriz identidade 10x10. A diferença entre o resultado
ideal e o resultado obtido é comparado utilizando "np.eye(10) - U_CGS.T @ U_CGS".

np.eye(10) é a matriz identidade 10x10.
U_CGS.T @ U_CGS é a matriz formada pelos produtos internos das colunas geradas pelo algoritmo
O resultado esperado deve ser próximo de 0.
"""
# a) Ortonormalização de uma base qualquer dada e teste do código em V
U_CGS = CGS(V)
print('Matriz gerada:')
print(U_CGS)

# b) Teste que verifica se a base obtida é de fato ortonormal
# Se o erro gerado for muito pequeno (<1e-12), é possível considerar que os vetores gerados são ortonormais.
err = np.linalg.norm(np.eye(10) - U_CGS.T @ U_CGS)
print(f'Erro do algoritmo CGS: {err}\n\n')
```

**Solução exercício 3**

O método do PCA é normalmente utilizado para visualizar diferentes grupos de um mesmo Dataset, em um número menor de componentes, de modo que os grupos fiquem separados. Isso é feito porque normalmente um único dado possui varios componentes que o descreve, como é o caso nesse dataset sobre vinhos. Cada amostra possui uma quantidade de alcool, ácido málico, ASH, magnesio, fenols, etc. E não é possível criar um eixo para cada uma dessas características e desenhar um ponto para cada vinho utilizando seus valores para cada característica. Isso é porque teriamos mais do que 3 eixos para demonstrar um único ponto e isso não é visualizavel para um humano. Dessa forma, é necessário utilizar algum algoritmo para reduzir a quantidade de eixo que visualizamos os pontos em um gráfico.

Para isso o PCA é um algoritmo excelente. Ele verifica quais dessas características possuem maior variação entre os dados, e utiliza somente esses eixos para demonstrar os gráficos. Além disso, selecionar os eixos com maior variância é um método interessante porque aumenta a probabilidade de elementos de diferentes classes estarem distantes entre sí, e dessa forma podemos formar *clusters* entre os dados e verificar qual grupo um novo ponto não classificado pode estar, somente baseado na sua projeção nesses eixos formados pelo PCA.

Agora que o propósito do PCA é claro, é possível entender como utiliza-lo:

Primeiro é necessário calcular a covariancia, dois a dois, entre cada uma das características das amostras, ou seja, calcular a covariancia entre: o alcool e o ácido málico, o ASH e o magnesio, etc. Se dizermos que X são nossos dados distribuidos da seguinte forma:

$$
X = 
\overbrace{
\begin{bmatrix}
-\!- & amostra\; 1 & -\!- \\
-\!- & amostra\; 2 & -\!- \\
 & \vdots & \\
-\!- & amostra\; n & -\!- \\
\end{bmatrix}
}^{m\; características}
$$

Podemos calcula uma matriz, com todas as covariâncias necessárias do seguinte modo:

$$
cov(X) =
\begin{bmatrix}
cov(alcool, alcool) & cov(alcool, ácido\;málico) & cov(alcool, ASH) & \dots & cov(alcool, fenols) \\
cov(ácido\;málico, alcool) & cov(ácido\;málico, ácido\;málico) & cov(ácido\;málico, ASH) & \dots & cov(ácido\;málico, fenols) \\
cov(ASH, alcool) & cov(ASH, ácido\;málico) & cov(ASH, ASH) & \dots & cov(ASH, fenols) \\
\vdots & \vdots & \vdots & \vdots & \vdots \\
\end{bmatrix}
$$

No qual, alcool representa os valores de alcool para cada uma das $n$ amostras da matriz X. De modo mais resumido, se $alcool$ for representado por $x_1$, $ácido\;málico$ for representado por $x_2$, etc. Temos:

$$
cov(X) =
\begin{bmatrix}
cov(x_1, x_2) & cov(x_1, x_3) & \dots & cov(x_1, x_n) \\
cov(x_2, x_1) & cov(x_2, x_3) & \dots & cov(x_2, x_n) \\
\vdots & \vdots & \ddots & \vdots \\
cov(x_n, x_1) & cov(x_n, x_2) & \dots & cov(x_n, x_n) \\
\end{bmatrix}
$$

O mais interessante desse processo, é que se os dados estiverem normalizados (os dados com média 0), essa matriz de covariância pode ser calculada de modo:

$$
cov(X) = \frac{1}{n}XX^\top
$$

Como ja dito antes, a ideia central do PCA é encontrar uma nova base, em que a covariância entre características diferente (por exemplo, $cov(alcool, acido\;málico)$) seja zero (ou próximo de 0), de modo que assim podemos tornar a exibição dos dados mais esparsa.

Já foi dito um teorema anteriormente que gera uma matriz diagonal a partir de uma transformação de base ($P$) e uma matriz diagonal ($D$), que foi chamado de Teorema Espectral. E é exatamente isso que é usado pelo algoritmos do PCA. Transformamos a matriz $cov(X)$ em uma matriz diagonal($D$) a partir de uma transformação de base ($P$). Isso é formalizado em:

$$
cov(X) = PDP^{-1}
$$

Então assim, podemos transformar a base original (formada pelas características das amostras observadas) em uma nova base (apartir de $P$ e $P^{-1}$) em que seus eixos possuam variâncias dadas pelos elementos da diagonal em $D$.

Mas até esse ponto, ainda temos o mesmo número de eixos, porém eixos diferentes (que não mais representam características da amostra). Podemos escolher quais são os vetores da nova base, que possuem maior variância em $D$ e projetar todos os dados da amostra somente nesses vetores. Essa projeção é feita do mesmo modo que foi utilizado no exercício anterior ao formar uma nova base ortonormal para uma dada base. 

Uma dúvida muito comum sobre esse algoritmo, é o por que de não verificarmos direto as características com maior variância e visualizamos os dados projetados somente nessas características. Bem, isso se da pelo fato de que podem existir outros eixos que possuem alta covariância com os eixos escolhidos, e a projeção, mesmo que utilizando os eixos com maior variância, tenderiam a ser não esparsos, pelo contrário, tenderiam à se agrupar em uma reta ou plano (dependendo se foi utilizado uma projeção em 2 ou 3 eixos).

```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA # Será permitido usar a função pronta para o PCA
```

```python
# lendo a base de dados
wine = load_wine()

print(wine['feature_names'])

# Dados originais armazenados em df_wine
df_wine = pd.DataFrame(wine['data'])

# Dados normalizados com o StandardScaller armazenados em df_wine_norm
df_wine_norm = pd.DataFrame(StandardScaler().fit_transform(df_wine))
```

```python
# a)
# Realizando o algoritmo do PCA para os dados não normalizados em df_wine.
# Numero de componentes utilizados: 2
df_wine_pca = pd.DataFrame(PCA(n_components=2).fit_transform(df_wine))

# Scatter plots das duas componentes geradas pelo PCA
# Diferenciadas por target. Classes presentes na legenda do gráfico.
plt.figure(figsize=(15,6), facecolor='white')
s = plt.scatter(df_wine_pca[0], df_wine_pca[1], c=wine['target'], cmap='viridis')
plt.xlabel('Componente 0', fontsize='large')
plt.ylabel('Componente 1', fontsize='large')
plt.title('Dados não normalizados')
plt.legend(*s.legend_elements(),loc="lower left", title="Classes")
plt.show()
```

```python
# b)
# Realizando o algoritmo do PCA para os dados normalizados em df_wine_norm.
# Numero de componentes utilizados: 2
df_wine_norm_pca = pd.DataFrame(PCA(n_components=2).fit_transform(df_wine_norm))

# Scatter plots das duas componentes geradas pelo PCA
# Diferenciadas por target. Classes presentes na legenda do gráfico.
plt.figure(figsize=(15,6), facecolor='white')
s = plt.scatter(df_wine_norm_pca[0], df_wine_norm_pca[1], c=wine['target'], cmap='viridis')
plt.xlabel('Componente 0', fontsize='large')
plt.ylabel('Componente 1', fontsize='large')
plt.title('Dados normalizados adequadamente')
plt.legend(*s.legend_elements(),loc="lower left", title="Classes")
plt.show()
```

**Diferença dos resultados:**

Olhando os gráficos gerados pelos dados normalizados, é possível ver uma explicita distinção entre os elementos de diferentes classes. Enquanto o resultado gerado pelo algoritmo PCA nos dados não normalizados é exatamente o oposto, visto que os dados não foram separados em grupos de forma adequada. Isso significa que o algoritmo do PCA não é adequado para decidir quais eixos possuem maior variância quando essas variâncias não são adequadamente normalizadas.

```python
# AVISO: Por prolemas de compatibilidade do plotly. O gráfico, mesmo que gerado no notebook,
# não é mantido quando aberto no Google Colab. Por isso, é necessário rodar essa parte do
# código ao utilizar o Google Colab ou o Jupyter Notebook.
# No Google Colab,:
# Clicar em "Runtime" > "Run All"(CTRL + F9) é o necessário para rodar todo o código sequencialmente.

# c)
# Utilizacao do plotly para plotar os dados em 3D
!pip install plotly # Instala o plotly caso o Google Colab não tenha o pacote previamente instalado
import plotly.express as px

wine_norm_pca_3 = PCA(n_components=3).fit(df_wine_norm)
df_wine_norm_pca_3 = pd.DataFrame(wine_norm_pca_3.transform(df_wine_norm))
df_wine_norm_pca_3.columns = ['Componente 0', 'Componente 1', 'Componente 2']
df_wine_norm_pca_3['target'] = wine['target']

fig = px.scatter_3d(df_wine_norm_pca_3, x='Componente 0', y='Componente 1', z='Componente 2', color='target')
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

**Variância explicada:**

Uma informaçao muito útil da técnica de aplicação do PCA é o calculo da informação que pode ser explicada por cada uma das componentes geradas. Esse fator é chamado de "variância explicada" do resultado e é dado por:

$$
\sigma^{exp} = \frac{\sigma_i}{\sum_{j}^{n} \sigma_{j}}
$$

Esse parametro é bastante interessante, ele representa o quanto os dados estão esparsos no eixo calculado, comparado com o quanto eles poderiam ser esparsos nos outros eixos. E se somarmos as variâncias explicadas dos eixos selecionados para realizar a projeção, temos uma noção do quanto que os eixos juntos deixam os dados esparsos.

Nesse ponto é importante ressaltar que quanto mais os dados são esparsos, maior a probabilidade de eles serem divididos em clusters. Então é importante que selecionemos a quantidade de clusters cujas variâncias explicadas sejam máximas. Mas ao mesmo tempo é importante lembrar que quanto maior a quantidade de eixos selecionados, maior a dificuldade de enxergar os dados, então essa seleção deve ser feita com cuidado.

Finalmente, como ja foi mencionado, é possível somar as variâncias explicadas de cada um dos componentes e verificar o quão esparso os dados estão comparados com o quão esparsos eles poderiam ficar (utilizando todos os eixos). Isso é formalizado em:

$$
\text{Variância explicada total} = \sigma^{exp}_{\text{Componente 0}} + \sigma^{exp}_{\text{Componente 1}} + \sigma^{exp}_{\text{Componente 2}}
$$

```python
## Exibir a variância explicada de cada um dos eixos separadamente.
for index, variance  in enumerate(wine_norm_pca_3.explained_variance_ratio_):
    print(f'Componente {index} = {variance * 100:.2f}%')

## Exibir a variância explicada quando utilizamos 3 componentes principais.
print(f'Soma das variâncias explicadas: {sum(wine_norm_pca_3.explained_variance_ratio_) * 100:.2f}%')
```

**Solução exercicio 4**

A ideia dentral dos mínimos quadrados, é estimar os parametros associados à uma funcao de forma que a função se assemelhe à distribuição de dados. O motivo principal para isso é querer aproximar novos valores (não presentes na distribuição de dados) de forma que eles sigam a *tendencia* do conjunto de dados observados. 

Para exemplificação, é possível pensar em uma função polinomial de grau máximo 1 ($ax + b$) que descreve bem certa distribuição de temperatura para cada hora do dia. Qual seriam os melhores parametros para $a$ e $b$ que levariam cada hora do dia ($x$) em determinada temperatura, de modo que a diferença do valor real (valor anotados para aquela hora do dia) e do valor estimado pela função seja mínima?

Ainda no exemplo da temperatura acima, se os dados já anotados fossem algo como:

\begin{array}{c|c}
\text{Hora} & \text{Temperatura} \\
\hline
1 & 21 \\
2 & 20 \\
3 & 19 \\
4 & 19 \\
5 & 19 \\
6 & 19 \\
7 & 20 \\
8 & 21 \\
9 & 22 \\
\end{array}

É possível testar alguns valores para os parametros $a$ e $b$ da equação linear, e comparar os valores finais com os resultados anotados pelas temperaturas da tabela acima. Como em:

$$
Valores\;testados: \\
a = 0\\
b = 20
$$

\begin{array}{c|c|c}
\text{Hora (x)} & f(x) = a * x + b & Temperatura - f(x)\\
\hline
1 & 0 * 1 + 20 = 20 & 21 - 20 = 1\\
2 & 0 * 2 + 20 = 20 & 20 - 20 = 0\\
3 & 0 * 3 + 20 = 20 & 19 - 20 = -1\\
4 & 0 * 4 + 20 = 20 & 19 - 20 = -1\\
5 & 0 * 5 + 20 = 20 & 19 - 20 = -1\\
6 & \vdots & \vdots\\
7 & 20 & 0\\
8 & 20 & 1\\
9 & 20 & 2\\
\end{array}

Nesse caso, como poderiamos fazer para medir o quão "errado" foi nossa predição? Um modo bastante efetivo é somar os erros individuais para ter uma visão geral com um único valor. Mas isso é inviável se obtermos valores negativos, visto que os erros podem se cancelar e no fim obtermos um erro igual a zero sendo que a predição feita não foi adequada. Para isso é de costume elevar os erros ao quadrado antes de realizar a soma. Além disso, existe uma estratégia chamada de "Erro quadratico médio" que se refere à essa soma dos quadrados, mas dividida pelo número de amostras nos dados (números de horas anotadas no exemplo acima), formalizado na seguinte forma:

$$
Erro = \frac{1}{n} \sum_{i}^{n} \left( y_i - f(x_i) \right)^2
$$

Infelizmente testar valores para os parametros e verificar quais quais erros são menores não é um processo eficiente. É possível utilizar algebra linear para resolver o problema de forma mais simples. Do mesmo modo que construimos a tabela, é possível formar um sistema linear com as equações:

$$
ax_1 + b \approx y_1 \\
ax_2 + b \approx y_2 \\
ax_3 + b \approx y_3 \\
\vdots \\
ax_n + b \approx y_n \\
$$

Essas equações acima querem dizer, que os parametros $a$ e $b$ devem aproximar todas as horas anotadas ($x$) nas temperaturas anotadas ($y$). O que podemos escrever, também, na forma matricial:

$$
\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots \\
1 & x_n \\
\end{bmatrix}
\begin{bmatrix}
a \\
b \\
\end{bmatrix}
\approx
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n \\
\end{bmatrix}
$$

E isso pode ser entendido como um sistema linear comum do tipo $Ax = b$. O principal problema dessa interpretação é que não podemos resolver um sistema linear que não seja quadrado com facilidade. Mas podemos transformar a matriz A em uma matriz quadrada multiplicando ambos os lados por $A^\top$, obtendo:

$$
A^\top A \alpha = A^\top y \\
$$

```python
# Funcao para calcular o erro quadrático médio
def erro_quadratico_medio(y, y_pred): return np.linalg.norm(y - y_pred)**2 / y.shape[0]

# Dados
x = np.array([1872, 1890, 1900, 1920, 1940, 1950, 1960, 1970, 1980, 1991, 2000, 2010])
y = np.array([9.9, 14.3, 17.4, 30.6, 41.2, 51.9, 70.9, 94.5, 121.1, 146.9, 169.5, 190.7])

# Exibição de todos os dados em um gráfico
plt.figure(figsize=(15,6), facecolor='white')
plt.plot(x, y, 'o-', linewidth=2)
plt.xlabel('Anos', fontsize='large')
plt.ylabel('População (milhões)', fontsize='large')
plt.title('População de 1872 até 2010')
plt.show()
```

```python
# Aproximação da curva por uma reta

"""
Primeiro é calculado a matriz A com os parametros x dos dados
elevados pelas respectivas potencias (nesse caso x^0 e x^1)
Essa matriz é amplamente conhecida, e chamada de Matriz de Vandermonde.
Por isso utilizamos o método implementado pelo numpy "numpy.vander()"
"""
A = np.vander(x, 2)
AtA = A.T @ A   # Calculo da matriz formada por AtA
Atb = A.T @ y   # Calculo da matriz formada por Aty
alpha = np.linalg.solve(AtA, Atb)   # Resolvendo sistema linear para os coeficientes do polinomio
y_pred = np.polyval(alpha, x)       # Obtendo os valores estimados com os coeficientes alphas

# Exibindo os gráficos originais e estimados para comparação.
plt.figure(figsize=(15,6), facecolor='white')
plt.plot(x, y, 'o-', label="Dados originais", linewidth=2)
plt.xlabel('Anos', fontsize='large') 
plt.ylabel('População (milhões)', fontsize='large') 
plt.title('População de 1872 até 2010')
plt.plot(x, y_pred, 'o-', label="Dados aproximados por uma reta", linewidth=2)
plt.xlabel('Anos', fontsize='large') 
plt.ylabel('População (milhões)', fontsize='large')

# Calculo do erro quadratico médio (Erro = 426.871 para uma aproximação por uma reta)
erro = erro_quadratico_medio(y, y_pred)
plt.title(f'Comparação dados populacionais com a aproximação por uma reta (Erro = {erro:.3f})')
plt.legend()
plt.show()
```

```python
# Aproximação da curva por uma parabola

"""
As funções utilizadas são semelhantes as funções utilizadas para gerar a aproximação linear acima.
A unica diferença está na hora de gerar a matriz de Vandermonde. Agora é utilizado o número 3 que representa
o conjunto dos polinomios de grau até 2.
"""
A = np.vander(x, 3)
AtA = A.T @ A
Atb = A.T @ y
alpha = np.linalg.solve(AtA, Atb)
y_pred = np.polyval(alpha, x)

plt.figure(figsize=(15,6), facecolor='white')
plt.plot(x, y, 'o-', label="Dados originais", linewidth=2)
plt.xlabel('Anos', fontsize='large') 
plt.ylabel('População (milhões)', fontsize='large') 
plt.title('População de 1872 até 2010')
plt.plot(x, y_pred, 'o-', label="Dados aproximados por uma parabola", linewidth=2)
plt.xlabel('Anos', fontsize='large') 
plt.ylabel('População (milhões)', fontsize='large')

# Calculo do erro quadrático médio para a aproximação por uma parabola (Erro = 20.776)
erro = erro_quadratico_medio(y, y_pred)
plt.title(f'Comparação dados populacionais com a aproximação por uma parabola (Erro = {erro:.3f})')
plt.legend()
plt.show()
```

```python
"""
Como o Erro com a utilização da aproximação por uma parabola (Erro = 20.776) é menor do que a aproximação por umar reta,
faz mais sentido calcular a predição da população em 2021 utilizando o método da aproximação por uma parabola.
"""
pop_2021 = np.polyval(alpha, 2021)  # Aproximando o valor de x = 2021 com os coeficientes da parabola em alpha.
print(f'Predição da população em 2021: {pop_2021:.3f} milhões de pessoas')
```
